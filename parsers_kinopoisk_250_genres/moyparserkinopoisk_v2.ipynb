```python
from bs4 import BeautifulSoup
import requests
```

```python
'''Код работал но опять начинает выдавать пустой массив'''
from time import sleep

allgenres = []
for i in range(1,6):
    sleep(3)
    url = f'https://www.kinopoisk.ru/lists/movies/series-top250/?page={i}'
    response = requests.get(url)
    bs = BeautifulSoup(response.text, "lxml")
    
    temp = bs.find_all('span', class_ = 'desktop-list-main-info_truncatedText__IMQRP')
    
    for x in temp:
        allgenres.append(x.text)
with open('C:\\Users\\Sammy\\Desktop\\info.txt','w') as file:
    allgenres_text = '\n'.join(allgenres)
    file.write(allgenres_text)
print(allgenres_text)
```

```python
import re
genres = {}
with open('C:\\Users\\Sammy\\Desktop\\info.txt','r') as file:
    rows = file.readlines()
    for row in rows:
        if 'В ролях' in row:
            continue
        c = row.split('•')[1].split("Режиссёр")[0]
        c1 = re.sub("[^А-Яа-я]", "", c)
        if c1 in genres:
            genres[c1] += 1
        else:
            genres[c1] = 1
print(genres)
for k, v in genres.items():
    print(k, v)
```
